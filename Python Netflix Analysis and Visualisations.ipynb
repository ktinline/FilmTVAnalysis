```python
#creating function to define movies/shows as excellent, good or poor depending on their imdb_score 

def score_function(x):
    if x >= 8.0: 
        return 'excellent'
    elif x >= 6.0 and x < 8.0: 
        return 'good'
    else: 
        return 'poor'
```

```python
df['score_category'] = df['imdb_score'].apply(score_function)
```
